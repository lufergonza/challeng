# Informe de Análisis de Desempeño de Tiendas

## Introducción

El presente informe tiene como objetivo analizar el desempeño de cuatro tiendas pertenecientes al Sr. Juan con el fin de determinar cuál de ellas debería vender. Para ello se evaluaron diferentes factores relevantes para el negocio, tales como los **ingresos totales**, las **categorías de productos más y menos vendidas**, las **calificaciones promedio de los clientes**, los **productos más y menos vendidos** y el **costo promedio de envío**.

A partir de estos indicadores y de los gráficos generados en el análisis, se busca identificar fortalezas y debilidades de cada tienda para tomar una decisión fundamentada sobre cuál tienda representa el menor desempeño dentro del conjunto.

---

# Desarrollo del análisis

## 1. Ingresos totales por tienda

Los ingresos totales obtenidos por cada tienda fueron los siguientes:

* **Tienda 1:** 1,150,880,400
* **Tienda 2:** 1,116,343,500
* **Tienda 3:** 1,116,343,500
* **Tienda 4:** 1,038,375,700

La **Tienda 1** presenta el mayor ingreso total, lo que indica un alto volumen de ventas y buen desempeño comercial. Por otro lado, **la Tienda 4 registra el ingreso más bajo**, lo que podría reflejar menor demanda o menor eficiencia en las ventas.

---

## 2. Categorías de productos más y menos vendidas

En general, las categorías con mayor volumen de ventas en las tiendas fueron:

* **Muebles**

Estas categorías concentran la mayor cantidad de productos vendidos en todas las tiendas, lo que indica que son los segmentos más importantes para el negocio.

Por otro lado, las categorías con menor cantidad de ventas suelen ser:

* **Instrumentos musicales**
* **Libros**
* **Artículos para el hogar**

Esto sugiere que estas categorías tienen una demanda menor en comparación con otras.
